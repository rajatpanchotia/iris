**PARALLEL COORDINATES**
        is a plotting technique for plotting multivariate data. Parallel coordinates allows one to see clusters in data and to estimate. other statistics visually. Using parallel coordinates points are represented as connected line segments. Each vertical line represents one attribute. One set of connected line segments represents one data point. Points that tend to cluster will appear closer together.

```
plt.figure()
parallel_coordinates(data1,"Species")
```

```
X=data1.iloc[:,:4]
Y=data1["Species"]
train_x, test_x, train_y, test_y = train_test_split(X,Y, test_size=0.3)
```

**LOGISTIC REGRESSION**


```
log = LogisticRegression(max_iter=1200,solver="lbfgs")
log.fit(train_x, train_y)
pred_lr = log.predict(test_x)
acc_lr = accuracy_score(test_y, pred_lr)
print(f"Accuracy of Logistic Regression Model: {round(acc_lr*100, 2)}%")
```

**DECISION TREE MODEL**

```
KNN = KNeighborsClassifier(n_neighbors=5)
KNN.fit(train_x, train_y)
pred_knn = KNN.predict(test_x)
acc_knn = accuracy_score(test_y, pred_knn)
print(f"Accuracy of KNN Model: {round(acc_knn*100, 2)}%")
```

**SVM MODEL**

```
svc = SVC()
svc.fit(train_x, train_y)
pred_svc = svc.predict(test_x)
acc_svc = accuracy_score(test_y, pred_svc)
print(f"Accuracy of SVC Model: {round(acc_knn*100, 2)}%")
```

**XG BOOST**

```
# Improving Accuracy using Boosting in Ml
from xgboost import XGBClassifier
import sklearn.metrics as metrics

xgb_classifier = XGBClassifier()
xgb_classifier.fit(train_x,train_y)
y_pred_xgb = xgb_classifier.predict(test_x)

acc_svc_xgb = metrics.accuracy_score(test_y, y_pred_xgb)
print(f"Accuracy after Boosting: {acc_svc_xgb*100}%")
```
